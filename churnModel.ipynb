# Data Preprocessing

```python
cid = data['CustomerID']
cid.head()
```

```python
# Drop the customerID column, reduce training features
data.drop(['CustomerID'], axis=1, inplace=True)
```

```python
data.replace({'No phone service': 'No', 'No internet service': 'No'}, inplace=True)
```

```python
data.head()
```

```python
X = data.drop(['Churn Label'], axis=1)
y = data['Churn Label']
```

```python
columns = []
for i in X.columns:
    if len(X[i].unique()) < 5:
        columns.append(i)
columns
```

```python
X = pd.get_dummies(X, columns=columns, drop_first=True)
X.head()
```

```python
X.drop(['Total Charges'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

columns = ['Tenure Months', 'Monthly Charges']
X_scaled = StandardScaler().fit_transform(X[columns])
X['Tenure Months'] = X_scaled[:, 0]
X['Monthly Charges'] = X_scaled[:, 1]
X.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 15))
sns.heatmap(X.corr(), annot=True, cmap='Blues')
```

### Observation


    .Here we see a strong correlation between InternetService_Fiber optic and MonthlyCharges which is obvious as people have fiber optic would be having a higher monthly charges than people without fiber optic.

    .A strong negative correlation is seen between InternetService_No and MonthlyCharges. The reason is intuitive.

    .Strong positive correlation is also seen between StreamingTV_Yes , StreamingMovies_Yes with MonthlyCharges. The reason is intuitive.


```python
y = pd.get_dummies(y, drop_first=True)
y
```

```python
sns.countplot(data=y, x='Yes')
```

##### Highly Imbalanced Dataset. We can do RandomOversampling

```python
from imblearn.over_sampling import RandomOverSampler

oversample = RandomOverSampler(sampling_strategy='minority')
X_over, y_over = oversample.fit_resample(X, y)
```

```python
X_over.shape
```

```python
y_over.shape
```

```python
sns.countplot(data=y_over, x='Yes')
```

```python
y_over = y_over['Yes']
```

```python
X_over.drop(['City', 'Lat Long','Churn Reason'], axis=1, inplace=True)
```

```python
X_over.shape
```

```python
y_over.shape
```

```python
X.drop(['City', 'Lat Long', 'Churn Reason'], axis=1, inplace=True)
```

```python
from sklearn.ensemble import ExtraTreesClassifier

# Apply the default seaborn them, scaling and color palette
sns.set()

# Setting up Multiple decision tree
extra_tree_forest = ExtraTreesClassifier(n_estimators=5, 
                                        criterion='entropy', 
                                        max_features= 2,
                                        random_state=42)

plt.figure(figsize=(20, 8))

extra_tree_forest.fit(X_over, y_over)

# Ensure that the model doesn't overfit the data
feature_importance = extra_tree_forest.feature_importances_

feature_importance_normalized = np.std([tree.feature_importances_ for tree in extra_tree_forest.estimators_], axis=0)

plt.bar(X.columns, feature_importance_normalized)
plt.xlabel('Feature Labels')
plt.ylabel('Feature Importances')
plt.xticks(rotation=90)
plt.title('Comparison of different Feature Importances')
plt.show()
```

# Model Building

```python
# Create Dependent and Independent Variables
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_over, y_over, random_state=42)
```

```python
len(X_train.columns)
```

# KNN Model

```python
from sklearn.svm import SVC 
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

knn_scores = []
for k in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5)
    knn_scores.append(scores.mean())

x_ticks = [i for i in range(1, 21)]
x_labels = x_ticks

plt.plot([k for k in range(1, 20)], knn_scores)
plt.xticks(ticks=x_ticks, labels=x_labels)
plt.grid()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

confusion_knn = confusion_matrix(y_test, knn.predict(X_test))
sns.heatmap(confusion_knn, annot=True)
print(classification_report(y_test, knn.predict(X_test)))
```
